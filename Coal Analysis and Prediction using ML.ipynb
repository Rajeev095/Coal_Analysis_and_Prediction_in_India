# Now, For Prediction...

```python
# Dropping Type of Mine
df.drop(['Type of Mine'],axis=1, inplace=True)
```

```python
df.head()
```

### _Reordering Columns_

```python
title = list(df.columns)
title
```

```python
title[0],title[2] = title[2],title[0]
title
```

```python
df=df[title]
```

```python
df.head()
```

## _One hot Encoding - Getting Dummy Variables_

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
df.head()
```

## _Independent and Dependent Features_

```python
# Production Column being the Dependent Feature..
X=df.iloc[:,1:]  #independent
y=df.iloc[:,0]
```

```python
y.head()
```

## _Extra trees Regressor Model_

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,y)
```

```python
#print(model.feature_importances_)
```

## _Train-Test Split_

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

# Using Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

### _Hyperparameter Tuning_

```python
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
print(n_estimators)
```

### _Using RandomizdSearchCV for hyperparameters_

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]

max_features = ['auto', 'sqrt']

max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]

min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10]
```

```python
## Selecting Parameters 

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
print(random_grid)
```

```python
rf = RandomForestRegressor()
```

```python
# Applying randomizedsearchcv

rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

# Fitting the Model

```python
rf_random.fit(X_train,y_train)
```

##### Looking for best parameters

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```

# Doing The Prediction of The Amount of Coal

```python
predict=rf_random.predict(X_test)
```

```python
predict
```

# _Now, Plotting the Prediction :_

```python
plt.figure(figsize=(13,8))
sns.distplot(y_test-predict, color ="maroon")
```

#### YAY!! We can see a Normal Distribution Curve 

---

### _Plotting the Best fit Line_

```python
plt.figure(figsize=(20,8))
sns.set_context('poster')
sns.regplot(y_test,predict, color ='teal')
```

## _Let's Now Look for the `Terrors`  Errors_ 

```python
from sklearn import metrics
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, predict))
```

```python
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, predict)))
```
